```python
from professor import *
from ratemyprof_api import *
```

```python
CornellUniversity = RateMyProfApi(298)
```

```python
cornell = CornellUniversity
```

```python
#cornell.search_professor("Walker White")
x = cornell.scrape_professors()
```

```python
profs = {}
for prof in x:
    profs[x[prof].name] = {'name': x[prof].name, 'num_ratings': x[prof].num_of_ratings, 'rating': x[prof].overall_rating}
```

```python
df = []

count = 0
for id_ in x:#[1705832, 2277575]: #1053274]:
#     if count >= 10:
#         break
    profID = x[id_].ratemyprof_id
    profName = x[id_].name
    profNumRatings = x[id_].num_of_ratings
    profOverallRating = x[id_].overall_rating
    #print("ID: ", x[id_].ratemyprof_id)
    #print("NAME: ", x[id_].name)
    #print("NUMBER OF RATINGS: ", x[id_].num_of_ratings)
    try:
        page = requests.get("https://www.ratemyprofessors.com/ShowRatings.jsp?tid=" + str(id_))

        y = page.content.decode('utf-8')
        avg = y.index('avgDifficulty')
        colon1 = y.index(':', avg + 1)
        end1 = y.index(',', avg + 1)
        takeagain = y.index('wouldTakeAgain')
        end2 = y.index(',', takeagain + 1)
        colon2 = y.index(':', takeagain + 1)

        avgDiff = y[colon1+1:end1]
        wouldTake = y[colon2+1:end2]
        
        #print(y[colon1 - 50: colon2+50])

        #print("AVERAGE DIFFICULTY: ", y[colon1+1:end1])
        #print("WOULD TAKE AGAIN: ", y[colon2+1:end2])
        df.append([profID, profName, profNumRatings, profOverallRating, avgDiff, wouldTake])
    except:
        #print('error')
        pass
    count += 1
```

```python
df2 = pd.DataFrame(df, columns = ['ID', 'Name', 'Number_of_Ratings', 'Average_Rating', 'Difficulty', 'Take Again'])
df2 = df2[df2.duplicated(subset="Name") == False]
df2.to_csv("RMP.csv", index = False)
```
