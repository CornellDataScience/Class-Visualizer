```python
import pandas as pd
import numpy as np
```

```python
class_roster = pd.read_csv("Classes_With_Medians.csv")
rmp_data = pd.read_csv("RMP/RMP.csv")
medians= pd.read_csv("Cornell_Median_Grades.csv")  
```

```python
semesterdict = {"FA21":0, 'SP21':1, 'FA20':2, 'SP20':3, 'FA19':4,'SP19':5,'FA18':6,'SP18':7,'FA17':8,'SP17':9,'FA16':10}
```

```python
#Data cleaning for medians
medians['Semester'] = medians['Semester (Ex: SP21)']
medians = medians.replace({"Semester": semesterdict})
medians = medians.sort_values(by="Semester")
medians = medians.drop_duplicates(subset="Dept + Number", keep="first")
#medians.to_csv("Cleaned_Medians.csv")
```

```python
class_roster=class_roster.drop(columns="Unnamed: 0")
```

```python
class_roster
```

```python
class_roster[class_roster['Course_Name']=='Problems in Byzantine Art']
```

```python
courses = []
for row in class_roster.to_numpy():
    course = row[0]+" " + str(row[1])
    courses.append(course)
```

```python
pre_coreqs = []
for row in class_roster.to_numpy():
    preco = row[-6] #normally  -6, if running again then -7
    new_list = []
    #print(type(preco))
    if type(preco) == str or not (np.isnan(preco)):
        
        for course in courses:
            if course in preco:
                new_list.append(course)
    else:
        new_list.append(np.nan)
    pre_coreqs.append(new_list)
#pre_coreqs
class_roster["Pre/Coreqs_Clean"] = pre_coreqs
```

```python
full_data = class_roster.merge(rmp_data, left_on='Professor', right_on ="Name", how="outer")
```

```python
full_data.shape
```

```python
full_data['Semester'] = full_data['Semester (Ex: SP21)']
full_data = full_data.replace({"Semester": semesterdict})
full_data = full_data.sort_values(by="Semester")
full_data = full_data.drop_duplicates(subset="Dept + Number", keep="first")
print(full_data.shape)
#full_data = full_data.drop_duplicates(subset="Course_Name", keep="first")
```
