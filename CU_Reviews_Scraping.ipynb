```python
import requests
import pandas as pd
```

```python
df = pd.read_csv('Full_Data_FA22.csv')
df.head()
```

```python
data = df.to_numpy()

url = 'https://www.cureviews.org/v2/getCourseByInfo/'

difficulty = []
ratings = []
workloads = []

for val in data:

    try:
        sub = val[1].lower()
        num = int(val[2])
        myobj = {'number': num, 'subject': sub}

        res = requests.post(url, json = myobj).json()['result']

        diff = res['classDifficulty']
        rating = res['classRating']
        workload = res['classWorkload']
        
    except:
        diff = None
        rating = None
        workload = None
    
    difficulty += [diff]
    ratings += [rating]
    workloads += [workload]
        
```

```python
len(workloads)
```

```python
df['CU_Reviews_Difficulty'] = difficulty
df['CU_Reviews_Rating'] = ratings
df['CU_Reviews_Workload'] = workloads
```
